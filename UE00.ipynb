```python
# In der ersten Code-Zelle verwenden wir Kommentare, weisen einer Variable einen Zahlenwert zu,
# quadrieren diesen Wert und speichern ihn in einer neuen Variable ab.
value_to_square = 42
squared_value = value_to_square**2
squared_value # Diese Zeile ist dazu da, den quadrierten Wert direkt im Notebook anzuzeigen.
```

```python
# Einer der einfachsten und meistgenutzten Datentypen in Python ist die Liste.

# Definieren einer neuen Liste:
new_list = [0, 1, 2, 3, 4, 5]

# Iterieren durch diese Liste und Aufrufen des 'print' Statements für jeden Wert in der Liste:
for value in new_list:
    # Merke: Einrückungen sind ein wesentlicher Teil der Programmiersprache und haben eine besondere Bedeutung!
    # Siehe bspw: https://www.w3schools.in/python-tutorial/concept-of-indentation-in-python/ für nähere Details.
    print(value)
```

```python
# Über einzelne Elemente der Liste können wir aber auch über eine sog. List Comprehension (https://www.w3schools.com/python/python_lists_comprehension.asp) iterieren.
[value for value in new_list]
```

```python
# Anhängen eines Elements an diese Liste
new_list.append(6)
```

```python
# Ausgeben der geraden Zahlen der Liste in reversed Order
[value for value in reversed(new_list) if value%2 == 0]
```

```python
# Natürlich können wir auch eigene Funktionen definieren, die, nach dem Ausführen dieser Zelle des Jupyter Notebook in allen anderen Zellen verwendet werden können.
def print_val_if_even(value):
    if value%2 == 0:
        print(value)
```

```python
# Verwenden der print_val_if_even() Funktion in einer List Comprehension:
# Einziger Unterschied: Diesmal werden die Liste und die darin enthaltenen Zahlen 0-6 durch Verwendung der vordefinierten Pythonfunktion range() generiert!

[print_val_if_even(value) for value in range(0, 7)]; # Der Strichpunkt am Ende des der Zeile unterdrückt den Output des Returnvalues der print()-Funktionsaufrufe im Jupyter Notebook
```

## UE00.a)
Generieren von Zufallszahlen mithilfe des NumPy [Generators](https://numpy.org/doc/stable/reference/random/index.html#random-quick-start).
<ol>
    <li>Initialisiere den Random Number Generator. Übergib <tt>seed=42</tt> um die Ergebnisse dieser Übung reproduzierbar zu machen.</li>
    <li>Generiere $n = 1000$ <strong>gleichverteilte</strong> aus einer diskreten Gleichverteilung $U_{[1,6]}$.<br>Speichere diese Zufallszahlen in der Variable <tt>dice_rolls</tt> ab und gib gib die ersten 10 generierten Ganzzahlen im Jupyter Notebook aus.</li>
    <li>Generiere $n = 1000$ <strong>normalverteilte</strong> Zufallszahlen aus $N(170,9.5^2)$.<br>Speichere diese Zufallszahlen in der Variable <tt>bodyheight</tt> und gib die ersten 10 generierten Zahlen im Jupyter Notebook aus.</li>
    <li>Erstelle den Ordner <tt>data</tt> im aktuellen Directory.</li>
    <li>Verwende die <a href="https://numpy.org/doc/stable/reference/generated/numpy.save.html">save()-Funktion</a> von NumPy um die Variablen <tt>bodyheight</tt> und <tt>dice_rolls</tt> in jeweils eigene Dateien (Dateinamen: <tt>bodyheight.npy</tt>, <tt>dice_rolls.npy</tt>) zur späteren Verwendung abzuspeichern. Achte dabei darauf, diese Dateien im eben erstellten Ordner <tt>data</tt> abzulegen. Ebenso soll das Keyword <tt>del</tt> dazu verwendet werden, die Variablen <tt>bodyheight</tt> und <tt>dice_rolls</tt> temporär zu löschen, um in UE01.b) das Laden von Informationen aus abgelegten Dateien üben zu können.</li>
</ol>

```python
# UE00.a)

# Random Number with seed 42
rng = np.random.default_rng(seed=42)

#Generate n=1000 Randomnumbers from U[1,6]
dice_rolls = rng.integers(low=1, high=7, size=1000)
print("First 10 dice-rolls:\n", dice_rolls[:10])

#1000 random nubers N(170, 9.5²)
bodyheight = rng.normal(loc=170, scale=9.5, size=1000)
print("First 10 Height:\n", bodyheight[:10])

np.save("data/bodyheight.npy", bodyheight)
np.save("data/dice_rolls.npy", dice_rolls)
del bodyheight, dice_rolls
```

## UE00.b)
Berechnen von Lage- und Streuparametern der generierten Stichproben in <tt>bodyheight.npy</tt> und <tt>dice_rolls.npy</tt>.
<ol>
    <li>Mittels der <a href="https://numpy.org/doc/stable/reference/generated/numpy.load.html#numpy.load">load()-Funktion</a> von NumPy laden wir nun die Dateien <tt>bodyheight.npy</tt> und <tt>dice_rolls.npy</tt> aus dem Ordner <tt>data</tt> wieder in die Variablen <tt>bodyheight</tt> und <tt>dice_rolls</tt>.</li>
    <li>Berechne Maximum, Minimum, Mittelwert, Standardabweichung und Median der ersten 500 Einträge in <tt>bodyheight</tt>. Gib die Ergebnisse <a href="https://docs.python.org/3/tutorial/inputoutput.html#tut-f-strings"><strong>formattiert</strong></a> und mit entsprechenden Labels im Jupyter Notebook aus.</li>
    <li>Berechne Mittelwert, Standardabweichung und Median der ersten 500 Einträge in <tt>dice_rolls</tt>. Achte dabei wiederum auf die korrekte Formattierung der Ausgaben.</li>
</ol>

```python
# UE00.b)
bodyheight = np.load("data/bodyheight.npy")
dice_rolls = np.load("data/dice_rolls.npy")

# Streuparameter first 500 bodyheight
bh = bodyheight[:500]
print("First 500  Bodyheight:")
print(f"Maximum:            {np.max(bh):.2f} cm")
print(f"Minimum:            {np.min(bh):.2f} cm")
print(f"Mittelwert:         {np.mean(bh):.2f} cm")
print(f"Standardabweichung: {np.std(bh):.2f} cm")
print(f"Median:             {np.median(bh):.2f} cm")

# Streuparameter first 500 dice_rolls
dr = dice_rolls[:500]
print("\nFirst 500 Dicerolls")
print(f"Mittelwert:         {np.mean(dr):.4f}")
print(f"Standardabweichung: {np.std(dr):.4f}")
print(f"Median:             {np.median(dr):.1f}")
```
